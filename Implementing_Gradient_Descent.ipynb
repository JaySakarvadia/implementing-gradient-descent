```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creation of data points
input = np.random.randn(10,1) #randomly formulated array of 10 x data points that fall on a normal distribution
output = 10 * input + np.random.rand() #randomly formulated array of 10 y data points that follow the y=mx+b form for interpretability
```

```python
#visualizing our data to show linear relationship
sns.scatterplot(x=input[:, 0], y=output[:, 0])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('relationship between x and y')
plt.show()
```

```python
weights = 0.0
biases = 0.0
function = weights * input + biases #this is the function that will be updated (weights and biases) to fit our data
learning_rate = 0.01
```

```python
def loss (output, function):
    sum_squared_error = 0

    for i in range(1, len(input)):
        sum_squared_error += ((output[i] - function[i])**2)

    MSE = sum_squared_error/len(input)

    return (MSE)
```

```python
def gradient_descent (input, output, weights, biases, learning_rate):
    dweights = 0
    dbiases = 0
    n = input.shape[0]

    for xi, yi in zip(input, output):
        dweights += (-2/n) * xi[0] * (yi - (weights * xi[0] + biases)) #derived from MSE equation with respect to weights
        dbiases += (-2/n) * (yi - (weights * xi[0] + biases)) #derived from MSE equation with respect to biases

    weights = weights - learning_rate * dweights
    biases = biases - learning_rate * dbiases

    return (weights, biases)
```

```python
for epoch in range (400):
    weights, biases = gradient_descent(input, output, weights, biases, learning_rate)
    function = weights * input + biases
    MSE = loss(output, function)
    print (f'Epoch: {epoch}     Loss: {MSE}     Weights: {weights}      Biases: {biases}')
```

```python
sns.scatterplot(x=input[:, 0], y=output[:, 0])
sns.lineplot(x=input[:, 0], y=function[:, 0], color='red')  # Line of best fit in red
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Scatter Plot of Input vs Output with Line of Best Fit')
plt.show()
```

# Gradient Descent for a Quadratic Function


```python
#creation of data points
input = np.random.randn(10,1) #randomly formulated array of 10 x data points that fall on a normal distribution
output = 10 * input**2 #randomly formulated array of 10 y data points that follow the y=ax^2 form for interpretability

sns.scatterplot(x=input[:, 0], y=output[:, 0])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('relationship between x and y')
plt.show()
```

```python
paramater1 = 0.0
function = paramater1 * input**2 #this is the function that will be updated (weights and biases) to fit our data
learning_rate = 0.01
```

```python
#working
def loss (output, function):
    sum_squared_error = 0

    for i in range(1, len(input)):
        sum_squared_error =+ ((output[i] - function[i])**2)

    MSE = sum_squared_error/len(input)

    return (MSE)
```

```python
#have to update dparamater1 += ...
def gradient_descent (input, output, paramater1, learning_rate):
    dparamater1 = 0
    n = input.shape[0]

    for xi, yi in zip(input, output):
        dparamater1 += (-2/n) * xi[0]**2 * (yi - paramater1 * xi[0]**2)

    paramater1 = paramater1 - learning_rate * dparamater1
    return (paramater1)
```

```python
for epoch in range (400):
    paramater1 = gradient_descent(input, output, paramater1, learning_rate)
    function = paramater1 * input**2
    MSE = loss(output, function)
    print (f'Epoch: {epoch}     Loss: {MSE}     Paramater1: {paramater1}')
```

```python
sns.scatterplot(x=input[:, 0], y=output[:, 0])
sns.lineplot(x=input[:, 0], y=function[:, 0], color='red')  # Line of best fit in red
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Scatter Plot of Input vs Output with Line of Best Fit')
plt.show()
```
